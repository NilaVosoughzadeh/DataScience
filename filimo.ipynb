```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
film = pd.read_csv("filimo.csv")
```

```python
film.info()
```

```python
film.head()
```

```python
Genres = list(film["Genre"])
```

```python
MovieGenres = []
for Genre in Genres:
  MovieGenres.append(Genre)

MovieGenresCount = {}

for MovieGenre in set(MovieGenres):
  MovieGenresCount[MovieGenre] = MovieGenre.count(MovieGenre)

MovieGenresCount
```

```python
Actors = list(film["Actors"])
```

```python
ActorNames = []
for i in Actors:
  ActorNames.append(i)

ActorsCount = {}

for ActorName in set(ActorNames):
  ActorsCount[ActorName] = ActorName.count(ActorName)

ActorsCount
```
