```python
tips = sns.load_dataset("tips")
print(tips.shape)
tips.head()
```

```python
print(tips["total_bill"].min())
print(tips["total_bill"].max())
sns.distplot(tips["total_bill"])
```

```python
print(tips["tip"].min())
print(tips["tip"].max())
sns.distplot(tips["tip"],bins=20,kde=False)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg") #xlim=(x1,x2) , ylim=(y1,y2)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,hue="sex")
```

```python
sns.lmplot(x="total_bill",y="tip",hue="sex",data=tips,col="sex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
```

```python
sns.pairplot(tips)
```

```python
tips.corr(numeric_only=True)
```

```python
sns.pairplot(tips,hue="sex")
```

```python
sns.countplot(x="day",data=tips,palette=sns.color_palette())
```

```python
sns.barplot(x="sex",y="tip",data=tips,palette=sns.color_palette("coolwarm"))
```

```python
sns.barplot(x="day",y="tip",data=tips,palette=sns.color_palette())
```

```python
sns.swarmplot(x="day",y="tip",data=tips,palette=sns.color_palette())
```
